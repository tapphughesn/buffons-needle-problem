# Buffon's Needle Problem Monte Carlo Simulation




$$ \LARGE p \approx \frac{\text{# crossings}}{\text{# needles}} \approx \frac{2l}{\pi d} $$




$$ \LARGE \Rightarrow \pi \approx \frac{2l(\text{# needles})}{d (\text{# crossings})} $$

```python
import numpy as np
from PIL import Image
from PIL import ImageDraw
import random as rnd
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 10
s = (500, 500)

im = Image.new('RGBA', s, (255,255,255,255))
draw = ImageDraw.Draw(im)

#draw ruled lines
for i in range(4):
    x1, y1 = 0, 100*(i+1)
    x2, y2 = 500, 100*(i+1)
    color = (0,0,0,230)
    draw.line(((x1,y1),(x2,y2)), fill=color, width = 2)

#draw needles and intersections
for i in range(N):
    y = rnd.random()*400 + 50
    x = rnd.random()*400 + 50
    alpha = rnd.random()*np.pi
    length = 50
    x1, y1 = (x - length*np.cos(alpha)/2, y - length*np.sin(alpha)/2)
    x2, y2 = (x + length*np.cos(alpha)/2, y + length*np.sin(alpha)/2)
    color = (0, 0, 255, 255)
    draw.line(((x1,y1),(x2,y2)), fill=color, width = 3)
        
plt.figure(figsize=(10,10))
plt.imshow(np.asarray(im),
           origin='lower')
plt.show()

print("number of needles dropped:", N)
```

```python
N = 100000
s = (500, 500)
num_crossings= 0

im = Image.new('RGBA', s, (255,255,255,255))
draw = ImageDraw.Draw(im)

#draw ruled lines
for i in range(4):
    x1, y1 = 0, 100*(i+1)
    x2, y2 = 500, 100*(i+1)
    color = (0,0,0,230)
    draw.line(((x1,y1),(x2,y2)), fill=color, width = 2)

#draw needles and intersections
for i in range(N):
    y = rnd.random()*400 + 50
    x = rnd.random()*400 + 50
    alpha = rnd.random()*np.pi
    length = 50
    x1, y1 = (x - length*np.cos(alpha)/2, y - length*np.sin(alpha)/2)
    x2, y2 = (x + length*np.cos(alpha)/2, y + length*np.sin(alpha)/2)
    color = (0, 0, 255, 255)
    draw.line(((x1,y1),(x2,y2)), fill=color, width = 3)
    #intersection pt
    if (int(y1/100) != int (y2/100)):
        num_crossings += 1
        argtop = np.argmax((y1,y2))
        top = np.asarray([(x1,y1), (x2,y2)])[argtop]
        y_cross = int(top[1]/100)*100
        # tan alpha = dy / dx
        x_cross = top[0] - (top[1] - y_cross) / np.tan(alpha)
        draw.line(((x_cross,y_cross-4),(x_cross,y_cross+4)),
                 fill = (255, 0, 0, 255), width = 8)
        
plt.figure(figsize=(10,10))
plt.imshow(np.asarray(im),
           origin='lower')
plt.show()

print("number of needles dropped:", N)
print("number of crossings:", num_crossings)
print("estimation of pi:", int(N *10000/ num_crossings)/10000)
print("error:", int(np.abs(np.pi - N/num_crossings)*100000/np.pi)/1000, "%")
```
